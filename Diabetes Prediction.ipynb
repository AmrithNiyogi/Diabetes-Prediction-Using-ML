```python
# importing the necessary libraries/dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a variable called data
data = pd.read_csv('diabetes.csv')
```

```python
# checking the first 10 values present in the data
data.head(10)
```

```python
#checking the last 10 values present in thed data
data.tail(10)
```

```python
# checking the shape of the data (number of rows and columns)
data.shape
```

```python
# getting a list of all the headings of the columns present in the data
data.columns.values.tolist()
```

```python
# getting the statistical measures of the data
data.describe(include='all')
```

```python
# checking the number of people who are diabetes positive/negative
# 0 -> negative
# 1 -> positive
data['Outcome'].value_counts()
```

```python
# group the data to check the mean based on the outcome
grouped_data = data.groupby('Outcome')
grouped_data.mean()
```

```python
# we separate the data and the labels
X = data.drop(columns="Outcome", axis=1) # we use axis=1 to specify to drop the columns
Y = data['Outcome']
```

```python
# printing the variables we created above
print(X)
```

```python
print(Y)
```

```python
# Data Standardization
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
# reassigning the standardized values
X = standardized_data
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the data into training and testing datasets
# test_size refers to the percentage of data we should exclude and use only for testing/validating the model
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# checking the values of the test and training variables assigned
print(f"Total Size: {X.shape}")
print(f"Total Training Size: {X_train.shape}")
print(f"Total Testing Size: {X_test.shape}")
```

```python
# training the model using our data
classifier = svm.SVC(kernel="linear")
```

```python
# training the svm classifier
classifier.fit(X_train, Y_train)
```

```python
# evaluating the model - accuracy score
X_train_prediction = classifier.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction, Y_train)
print(f'Accuracy Score of training data: {X_train_accuracy}')
```

```python
# evaluating the accuracy of the testing data
X_test_prediction = classifier.predict(X_test)
X_test_accuracy = accuracy_score(X_test_prediction, Y_test)
print(f'Accuracy Score of testing data: {X_test_accuracy}')
```

```python
# building a predictive system
# input_data = input("Enter the data in the following order: Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigree,Age")
input_data = (9,119,80,35,0,29,0.263,29)
# changing the type to numpy array
data_array = np.array(input_data)
#reshape the data
input_data_reshaped = data_array.reshape(1,-1)

#standardizing the data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
if prediction[0] == 0:
    print("Non-Diabetic")
else:
    print("Diabetic")
```
